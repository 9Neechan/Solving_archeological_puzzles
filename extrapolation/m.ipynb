```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("many_pieces.jpg")
#image = cv2.imread("piece.PNG")

img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.axis('off')
plt.imshow(img)
```

```python
plt.axis('off')
plt.imshow(gray)
```

### Erosion

```python
for i in range(0, 3):
	eroded = cv2.erode(gray.copy(), None, iterations=i + 1)
	plt.axis('off')
	plt.imshow(eroded)
```

### Dilate

```python
for i in range(0, 3):
	dilated = cv2.dilate(gray.copy(), None, iterations=i + 1)
	plt.axis('off')
	plt.imshow(dilated)
```

### Morphological gradient

```python
kernelSize = (100, 100)
# loop over the kernels a final time

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, kernelSize)
k2 = cv2.getStructuringElement(cv2.MORPH_RECT, (13, 5))
gradient = cv2.morphologyEx(gray, cv2.MORPH_GRADIENT, k2)
```

```python
plt.axis('off')
plt.imshow(gradient)
```
